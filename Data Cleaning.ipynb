## Encounters data

```python
encounters = pd.read_csv('encounters.csv')

encounters.head()
```

## Diagnosis data

```python
diagnosis = pd.read_csv('diagnosis.csv')

diagnosis.head()
```

`GroupName` and `GroupCode` are broader categories to look into. 

```python
diagnosis['GroupName'].str.contains('fracture', case=False).mean() # Fractures only make up 27% of diagnoses

# Dental, diabetes, burns, open wounds are still a much lower percentage than Fractures
```

### Encounters x diagnosis data 

```python
df = pd.merge(encounters, diagnosis, left_on='PrimaryDiagnosisKey', right_on='DiagnosisKey')
df.head()
```

```python
df['GroupName'].value_counts().head(10)
```

Encounter for screening for malignant neoplasms could be our focus.

```python
list(df['GroupName'].unique())
```

```python
list(df.loc[df['GroupName'].str.contains('malignant', case=False, na=False), 'GroupName'].unique())
```

### What percentage of encounters are fractures/cancer?

```python
df['FRACTURE'] = df['GroupName'].str.contains('fracture', case=False, na=False).astype(int)
df['CANCER'] = df['GroupName'].str.contains('malignant', case=False, na=False).astype(int)
```

```python
df['FRACTURE'].mean() * 100
```

```python
df['CANCER'].mean() * 100
```

```python
(df['GroupName'] == 'Encounter for general examination without complaint, suspected or reported diagnosis').mean() * 100

# From this, we can see that although percentages of fracture and cancer are low, they are not as low when compared to the most common diagnosis.
```
